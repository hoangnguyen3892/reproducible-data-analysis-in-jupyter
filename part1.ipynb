````python
plt.style.use('ggplot') #instead of importing seaborn, we can use this one
data.columns = ['West', 'East'] #Change and shorten name of columns
data.resample('W').sum().plot();
````

````python
#plt.style.available
````

````python
# Zoom-in by rolling the window with 365-day width
data.resample('D').sum().rolling(365).sum().plot();
````

````python
# Set value for y-stick
data['Total'] = data['West'] + data['East']
ax = data.resample('D').sum().rolling(365).sum().plot();
ax.set_ylim(0, None)
````

````python
# Group by the data through the time of a day
data.groupby(data.index.time).mean().plot()
````

````python
len(data.index.time)
````

Now, in order to look into the trend through out time daily, we can use ```pivot_table``` to create a new dataframe which columns are days and rows are time of a day

[```pandas.pivot_table```](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.pivot_table.html#pandas-pivot-table): Create a spreadsheet-style pivot table as a DataFrame. 

````python
pivoted = data.pivot_table('Total', index=data.index.time, columns=data.index.date)
````

````python
pivoted.iloc[:5, :5]
````

````python
pivoted.plot(legend=False)
````

This plot is not really visible to observe the trend and find the hypothesis. It'd be better to make the lines become transparent. In order to achieve this, we pass the argument ```alpha``` into the function.

````python
pivoted.plot(legend=False, alpha=0.01)
````

**Note**: As can be seen from the figures, the majority of cases hitting the 'peak' twice a day and number of bikes of the remain are higher during mid-day. This can leads to the hypothesis that the former is on weekday while the later is at weekend.
